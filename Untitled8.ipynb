```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
data = {
    "advertising": [10, 15, 20, 25, 30, 35, 40, 45, 50, 55],
    "sales": [12, 18, 25, 28, 35, 40, 45, 50, 55, 60]
}

# Create dataframe
df = pd.DataFrame(data)

# Input and output
X = df[["advertising"]]
y = df["sales"]

# Train model
model = LinearRegression()
model.fit(X, y)

# Prediction
prediction = model.predict([[60]])

print("Predicted Sales for advertising 60 =", prediction[0])

# Graph
plt.scatter(df["advertising"], df["sales"])

# Regression line
plt.plot(
    df["advertising"],
    model.predict(X)
)

# Labels
plt.xlabel("Advertising Budget")
plt.ylabel("Sales")
plt.title("Sales Prediction Project")

# Show graph
plt.show()
```
